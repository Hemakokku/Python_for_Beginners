```python
plt.bar([0.25,1.25,2.25,3.25,4.25],[50,40,70,80,20],
label='bmw',width=.5)
plt.bar([.75,1.75,2.75,3.75,4.75],[80,20,20,50,60],
label='audi',color='r',width=.5)
plt.legend()
plt.xlabel('days')
plt.ylabel('distance (kms)')
plt.title('information')
plt.show()
```

```python
year=[2015,2016,2017,2018,2019,2020,2021,2022,2023,2024]
width=0.25
indices=np.arrange(len(year))
population_china=[1.359,1.397,1.403,1.409,1.415,1.42,1.424,1.428,1.431,1.434]
population_India=[1.23,1.309,1.324,1.339,1.354,1.368,1.397,1.411,1.425,1.455]
plt.bar(indices-width,population_china,width=width,label='china',color='g')
plt.bar(indices-width,population_India,width=width,label='India',color='r')
plt.title('population comparison between India and china')
plt.xlabel('year')
plt.ylabel('population in billions')
plt.legend()
plt.xticks(ticks=indices,labels=year)

plt.tight_layout()
plt.grid(True)
plt.show()
```

# Histogram plot

```python
fig,ax=plt.subplots(1,1)
a= np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
ax.hist(a, bins=[0,25,50,75,100])
ax.set_title("histogram of result")
ax.set_xticks([0,25,50,75,100])
ax.set_xlabel('marks')
ax.set_ylabel('no. of students')
plt.show()
```

# pie plot

```python
slices=[1433783686,1366417754,329006491,320625568,216565318]
labels=['china','india','united states','indonesia','pakistan']
explode=[0,0,0.2,0,0]
colors=['#c70039','#ff5733','#ffc300','#daf7a6']
plt.pie(slices,labels=labels,explode=explode,shadow=True,startangle=90,wedgeprops={'edgecolor':'black','linewidth':1},colors=colors)

plt.title('population pie chart')
plt.grid(True)
plt.show()
```

# scatter plot

```python
x=[1,1.5,2,2.5,3,3.5,3.6]
y=[7.5,8,8.5,9,9.5,10,10.5]

x1=[8,8.5,9,9.5,10,10.5,11]
y1=[3,3,5,3.7,4,4.5,5.2]


plt.scatter(x,y, label='high income low saving',color='r')
plt.scatter(x1,y1, label='low income high saving',color='b')
plt.xlabel('saving*100')
plt.ylabel('income*1000')
plt.title('scatter plot')
plt.legend()
plt.show()

```

```python
girls_grades=[80,98,90,90,88,77,88,99]
boys_grades=[33,77,88,99,66,55,44,76]
grades_range=[10,20,30,40,50,60,70,80]
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(grades_range,girls_grades,color='r')
ax.scatter(grades_range,boys_grades,color='b')
ax.set_xlabel('grades range')
ax.set_ylabel('grades scored')
ax.set_title('scatter plot')
plt.show()
```

# violin plot

```python
collection_1=np.random.normal(100,10,200)
collection_2=np.random.normal(99,30,200)
collection_3=np.random.normal(90,20,200)
collection_4=np.random.normal(80,40,200)


##combine these different collections into a list
data_to_plot=[collection_1,collection_2,collection_3,collection_4]
#create a figure instance
fig=plt.figure()
#create an axes instance
ax=fig.add_axes([0,0,1,1])
#create the boxplot
bp=ax.violinplot(data_to_plot)
plt.show()
```
