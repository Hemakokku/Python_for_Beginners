# Plotting univariate distribution with distplot()

the most important thing to do while analysing the data is to understand its distribution.seaborn helps us to understand the univariate distribution of data.
the distplot() function provides a quick look at univariate distribution.this function will plot a histogram that fits the kernel density estimates of the data.
we can use the distplot() funcy=tion as follows:-


```python
sns.distplot(tips['total_bill'])
plt.show()
```

# Seaborn Histogram

histogram represents the data distribution by forming bins along the range of the data and then drawing bars to show the number of observations that fall in each born.

we can use the same distplot() function to plot a histogram as follows:-


```python
sns.distplot(tips['total_bill'], kde=False)

plt.show()
```

# seaborn-kernel density estimates


kernel density estimation (kde) is a way to estimate  the probability density function of a continuous random variable.it is used for non-parametric analysis.

setting the hist parameter to false in distplot() function will yield the kernel density estimation plot.

```python
sns.distplot(tips['total_bill'],hist=False)
plt.show()
```

# plotting bivariate distribution with jointplot()


bivariate distribution is used to determine the relation between two variables.this mainly deals with relationship between two variables and how one variable is behaving with respect to the other.

the best way to analyze bivariate distribution in seaborn is by using the joinplot() function.

joinplot() creates a multi-panel figure that projects the bivariate relationship between two variables and also the univariante distribution of each variable on separate axes.

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
plt.show()
```

# seaborn-scatter plot

A scatter plot can be used to demonstraterelationship between two variables x and y.a simple scatter plot can be drawn as follows:-

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.show()
```

# Visualizing pairwise relationship with pairplot() 
some datasets contain many variables in such cases the relationship between each and every variable should be analyzed.so, we need to plot pairwise relationships in a dataset.
to plot multiple pairwise distributions,we can use pairplot() function.this shows the relationship for (n,2) combinationof variable in a dataframe as a matrix of plots and te diagonal plots are the univariate plots.

we can plot a pairplot as follows:-




```python
sns.set_style('ticks')
sns.pairplot(tips)
plt.show()
```

# seaborn-strip plot

a stripplot() is used to draw a scatterplot where one variable is categorical. it represents the data in sorted order any one of the axis.
we can plot stripplot as follows:- 

```python
sns.stripplot(x='day', y='total_bill', data=tips)
plt.show()
```

we can add jitter to bring out the distribution of values as follows

```python
sns.stripplot(x='day', y='total_bill', data=tips,jitter=True)
plt.show()
```

# seaborn-swarm plot

another option which can be used as an alternate to jitter is to use function swarmplot(). this fun positions each point of scatter plot on the categorical axis and thereby avoids overlapping points.

so swarmplot() can be used to draw categorical scatterplot with non-overlapping points.

we can plot swarmplot as follows:-

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
plt.show()
```

# seaborn-box plot


we can draw a bix plotto show distributions with respect to categories.boxplot is a convenient way to visualize the distribution of data through their quartiles.

box plots usually have vertical lines extending from the boxes which are termed as whiskers. these whiskers indicate variability outside the upper and lower quartiles.so box plots are also termed as box-and-whisker plot.any outliers in the data are plotted as individual points.

```python
sns.boxplot(x=tips['total_bill'])
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.show()
```

# seaborn-violin plot


violin plots are combinations of the box plot with the kernel density estimates.so,these plots are easier to analyze and understand the distribution of the data.

we can draw a single horizontal violin plot as follows:-

```python
sns.violinplot(x=tips['total_bill'])
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
plt.show()
```

# seaborn-bar plot

barplot sows the relation between a categorical var and a continuous var.the data is represented in rectangular bars where the length te bar represents the proportion of the data in that category.

a barplot show point estimates and confidence intervals as rectangular bars. so, it represents the estimate of central tendency.

we can draw a set of vertical bar plot grouped by a categorical var as follows:-


```python
sns.barplot(x='day',y='total_bill',data=tips)
plt.show()
```

# seaborn-point plot

```python
sns.pointplot(x='time',y='total_bill',data=tips)
plt.show()
```

```python
sns.pointplot(x='time',y='total_bill', hue='smoker',data=tips)
plt.show()
```

# seaborn-reg plot

```python
sns.regplot(x='total_bill',y='tip',data=tips)
plt.show()
```
